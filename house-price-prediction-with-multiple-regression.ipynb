# <span style="color:purple;">**House Price Prediction with Multiple Linear Regression**

# <span style="color:purple;">**Introduction**

#### **Linear regression is a fundamental supervised learning algorithm in machine learning. It aims to establish a linear relationship between a dependent variable (target) and one or more independent variables (features). In the context of house price prediction, the dependent variable will be the MEDV, and the independent variables can be factors like CRIM, ZN, INDUS, CHAS, NX, RM, AGE, DIS, RAD, TAX, PTRATIO, B, LSTAT.**

### <span style="color:purple;"> **Scenario**
    
#### **To predicting house prices using linear regression for real estate.**

```python
# step 1. import library
import pandas as pd
```

```python
# step 2. read data
df= pd.read_csv('https://raw.githubusercontent.com/YBIFoundation/Dataset/main/Boston.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# step 3. Define y and x
df.columns
```

```python
y= df['MEDV']
x= df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
```

```python
#step 4. Train test split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size= 0.7,random_state= 2529)
x_train
```

```python
# check shape of train and test sample
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
# step 5. Select Model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# step 6. Train or Fit Model
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# step 7. Predict model
y_pred = model.predict(x_test)
y_pred
```

```python
# step 8. Model Accuracy
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_pred)
```
